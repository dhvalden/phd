```python
import pandas as pd
```

```python
df = pd.read_csv('data/climate_sent.json.csv')
df['date'] = pd.to_datetime(df['date'])
df = df.set_index('date')
df_mean = df.iloc[:, 0:4].resample("1D").mean()
df_count = df.iloc[:, 4:].resample("1D").mean()
df_mean[['neg_mean', 'pos_mean']].plot()
```

```python
events = pd.read_csv('data/xr_events.csv')
events['date'] = pd.to_datetime(events.event_date)
events = events.set_index('date')
events_count = events['data_id'].resample("1D").count()
events_count = events_count['2018-12-05': '2020-02-01'] #keeping only relevant time frame
events_count.plot()
```

```python
merge = pd.merge(df_mean, events_count, how='inner', left_index=True, right_index=True)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 8]
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(merge.index, merge.compound_mean, 'g-')
ax2.plot(merge.index, merge.data_id, 'b-')
plt.show()
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(merge.index, merge.neg_mean, 'g-')
ax2.plot(merge.index, merge.data_id, 'b-')
plt.show()
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(merge.index, merge.pos_mean, 'g-')
ax2.plot(merge.index, merge.data_id, 'b-')
plt.show()
```

```python
idx = merge.loc[merge['data_id'] >= 10].index
#idx = idx + pd.DateOffset(days=-15)
idx = merge.index.isin(idx)
```

```python
def mean_sent_by_offset(data, offset, ra_window, eventmag=1):
    tempdata = data.rolling(ra_window, win_type='gaussian').mean(std=3)
    idx_0 = tempdata.loc[tempdata['data_id'] >= eventmag].index
    days_offset = []
    mean_result = []
    for i in range(0 - offset, offset + 1):
        idx = idx_0 + pd.DateOffset(days=i)
        idx = data.index.isin(idx)
        output = tempdata[idx].mean().to_dict()
        days_offset.append(i)
        mean_result.append(output)
    return pd.DataFrame.from_records(mean_result, index=days_offset)
```

```python
results = mean_sent_by_offset(merge, 10, 2, 10)
results.neg_mean.plot()
```
